```python
# code here
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

1) Aplique nas imagens abaixo o algortimo Watershed para realizar a contagem de células. Faça tanto para  a imagem original e para a imagem gradiente

```python
img1 = cv2.imread("image.jpg", cv2.IMREAD_GRAYSCALE)

# Aplique um filtro de suavização para reduzir ruídos
imagem_suavizada = cv2.medianBlur(img1, 5)

# Aplique a detecção de bordas para obter a imagem gradiente
gradiente = cv2.morphologyEx(imagem_suavizada, cv2.MORPH_GRADIENT, np.ones((3, 3), np.uint8))

# Aplique a transformada de distância para determinar as regiões de células
_, marcadores = cv2.connectedComponents(cv2.threshold(gradiente, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1])
_, marcadores2 = cv2.connectedComponents(cv2.threshold(img1, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1])

# Aplique o algoritmo Watershed para separar as células próximas
marcadores[gradiente == 255] = 1
marcadores = cv2.watershed(cv2.cvtColor(img1, cv2.COLOR_GRAY2BGR), marcadores)

marcadores2[img1 == 0] = 1
marcadores2 = cv2.watershed(cv2.cvtColor(img1, cv2.COLOR_GRAY2BGR), marcadores2)

# Destaque as regiões de células na imagem colorida
img_color = cv2.cvtColor(img1, cv2.COLOR_GRAY2BGR)
img_color[marcadores == -1] = [255, 0, 0]# Definir as bordas das células como vermelho



# Conte o número de células
numero_de_celulas = len(np.unique(marcadores)) - 1
numero_de_celulas2 = len(np.unique(marcadores2)) - 1

# Exiba os resultados
plt.subplot(1, 2, 1)
plt.imshow(img_color)
plt.title("Imagem original com células destacadas")
plt.subplot(1, 2, 2)
plt.imshow(gradiente, cmap='gray', vmin=0, vmax=255)
plt.title("Imagem gradiente")
plt.show()

print("Número de células encontradas:", numero_de_celulas)
print("Número de células encontradas:", numero_de_celulas2)
```
