2)	Utilizando a imagem ctskull-256.tif, refaça a quantização de seus pixels utilizando de 7 a 1 bit(s) por pixel.  Fazer os cálculos por meio do seu algortimo. Não usar o opencv para a conversão

```python
import numpy as np
import matplotlib.pyplot as plt

def resample_image(img, new_dpi, original_dpi):
    scale_factor = original_dpi / new_dpi
    new_shape = (int(img.shape[0] / scale_factor), int(img.shape[1] / scale_factor))
    new_img = np.zeros(new_shape, dtype=np.uint8)

    for i in range(new_shape[0]):
        for j in range(new_shape[1]):
            x = int(j * scale_factor)
            y = int(i * scale_factor)
            new_img[i, j] = img[y, x]

    return new_img

# Carrega a imagem original
img = plt.imread("ctskull-256.tif")

# Define os números de bits desejados
num_bits = [7, 6, 5, 4, 3, 2, 1]

# Realiza a quantização para cada número de bits
for bits in num_bits:
    new_img = np.zeros_like(img)
    # Realiza a quantização
    levels = 2**bits
    new_img = (img // (256 // levels)) * (256 // levels)
    # Plota a imagem quantizada
    plt.imshow(new_img, cmap='gray', vmin=0, vmax=255)
    plt.title(f"{bits} bits")
    plt.show()
```
